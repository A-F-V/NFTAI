#VAE

```python
class Encoder(nn.Module):
    def __init__(self):
        super(Encoder, self).__init__()

        self.mapping_dim = 10


        self.conv_dim = 100


        self.conv = nn.Sequential(
            nn.Conv2d(1,self.conv_dim, kernel_size=3), #30
            nn.LayerNorm([30,30]),
            nn.LeakyReLU(0.2),

            nn.Conv2d(self.conv_dim,self.conv_dim, kernel_size=3), #28
            nn.LayerNorm([28,28]),

            nn.LeakyReLU(0.2),

            nn.Conv2d(self.conv_dim,self.conv_dim, kernel_size=3), #26
            nn.LayerNorm([26,26]),
 
            nn.LeakyReLU(0.2),

            nn.Conv2d(self.conv_dim,self.conv_dim, kernel_size=3), #24
            nn.LayerNorm([24,24]),
      
            nn.LeakyReLU(0.2),

            nn.Conv2d(self.conv_dim,self.conv_dim, kernel_size=3), #22
            nn.LayerNorm([22,22]),
      
            nn.LeakyReLU(0.2),

            nn.Conv2d(self.conv_dim,self.conv_dim, kernel_size=3), #20
            nn.LayerNorm([20,20]),
  
            nn.LeakyReLU(0.2),

            nn.Conv2d(self.conv_dim,self.conv_dim, kernel_size=3), #18
            nn.LayerNorm([18,18]),
        
            nn.LeakyReLU(0.2),

            nn.Conv2d(self.conv_dim,self.conv_dim, kernel_size=3), #16
            nn.LayerNorm([16,16]),
     
            nn.LeakyReLU(0.2),

            nn.Conv2d(self.conv_dim,self.conv_dim, kernel_size=3), #14
            nn.LayerNorm([14,14]),
     
            nn.LeakyReLU(0.2),

            nn.Conv2d(self.conv_dim,self.conv_dim, kernel_size=3), #12
            nn.LayerNorm([12,12]),
  
            nn.LeakyReLU(0.2),

            nn.Conv2d(self.conv_dim,2, kernel_size=3,stride=2), #5
            nn.LayerNorm([5,5]),
            nn.LeakyReLU(0.2),
        )

        self.linear_mu = nn.Sequential(
            nn.Flatten(),
            nn.Linear(50,50),
     
            nn.LeakyReLU(0.2),
        )

        self.linear_var = nn.Sequential(
            nn.Flatten(),
            nn.Linear(50,50),
     
            nn.LeakyReLU(0.2),
        )

    def forward(self, img):
        x1 = self.conv(img)
        mu = self.linear_mu(x1)
        var = self.linear_var(x1)
        

        return mu, var
```

```python
class Decoder(nn.Module):
    def __init__(self):
        super(Decoder, self).__init__()

        self.mapping_dim = 10


        self.conv_dim = 200


        self.conv = nn.Sequential(
            nn.ConvTranspose2d(50,self.conv_dim, kernel_size=3),
            nn.LayerNorm([3,3]),
          
            nn.LeakyReLU(0.2),

            nn.ConvTranspose2d(self.conv_dim,self.conv_dim, kernel_size=5),
            nn.LayerNorm([7,7]),
      
            nn.LeakyReLU(0.2),

            nn.ConvTranspose2d(self.conv_dim,self.conv_dim, kernel_size=5),
            nn.LayerNorm([11,11]),
        
            nn.LeakyReLU(0.2),

            nn.ConvTranspose2d(self.conv_dim,self.conv_dim, kernel_size=5),
            nn.LayerNorm([15,15]),
      
            nn.LeakyReLU(0.2),

            nn.ConvTranspose2d(self.conv_dim,self.conv_dim, kernel_size=5),
            nn.LayerNorm([19,19]),
        
            nn.LeakyReLU(0.2),

            nn.ConvTranspose2d(self.conv_dim,self.conv_dim, kernel_size=5),
            nn.LayerNorm([23,23]),
      
            nn.LeakyReLU(0.2),

            nn.ConvTranspose2d(self.conv_dim,self.conv_dim, kernel_size=5),
            nn.LayerNorm([27,27]),
         
            nn.LeakyReLU(0.2),

            nn.ConvTranspose2d(self.conv_dim,self.conv_dim, kernel_size=4),
            nn.LayerNorm([30,30]),
        
            nn.LeakyReLU(0.2),


            nn.ConvTranspose2d(self.conv_dim,self.conv_dim, kernel_size=3),
            nn.LayerNorm([32,32]),

            nn.LeakyReLU(0.2),

            nn.Conv2d(self.conv_dim,self.conv_dim, kernel_size=3,padding=1,padding_mode='replicate'),
            nn.LayerNorm([32,32]),

            nn.LeakyReLU(0.2),
            

            nn.Conv2d(self.conv_dim,1, kernel_size=3,padding=1,padding_mode='replicate'),
            nn.Tanh(),
        )

    def forward(self, img):
        x1 = self.conv(img[:,:,None,None])
        

        return x1
```

```python
encoder = Encoder()
decoder = Decoder()
encoder.cuda()
decoder.cuda()
lr_encoder=0.0001
lr_decoder=0.0001

optimizer_encoder = torch.optim.Adam(encoder.parameters(), lr=lr_encoder)
optimizer_decoder = torch.optim.Adam(decoder.parameters(), lr=lr_decoder)
def gaussian_likelihood(x_hat, logscale, x):
        scale = torch.exp(logscale)
        mean = x_hat
        dist = torch.distributions.Normal(mean, scale)

        # measure prob of seeing image under p(x|z)
        log_pxz = dist.log_prob(x)
        return log_pxz.sum(dim=(1, 2, 3))

def kl_divergence(z, mu, std):
        # --------------------------
        # Monte carlo KL divergence
        # --------------------------
        # 1. define the first two probabilities (in this case Normal for both)
        p = torch.distributions.Normal(torch.zeros_like(mu), torch.ones_like(std))
        q = torch.distributions.Normal(mu, std)

        # 2. get the probabilities from the equation
        log_qzx = q.log_prob(z)
        log_pz = p.log_prob(z)

        # kl
        kl = (log_qzx - log_pz)
        kl = kl.sum(-1)
        return kl
```

# Train

```python
import matplotlib.pyplot as plt
start = False
n_epochs =int(1000000000)

for epoch in range(n_epochs):

    for i, (imgs) in enumerate(dataloader):

        imgs = imgs[0].cuda()

        optimizer_encoder.zero_grad()
        optimizer_decoder.zero_grad()

        # encode x to get the mu and variance parameters
        mu, log_var = encoder(imgs)
        # sample z from q
        std = torch.exp(log_var / 2)
        q = torch.distributions.Normal(mu, std)
        z = q.rsample()
        # decoded
        x_hat = decoder(z)
        # reconstruction loss
        log_scale = nn.Parameter(torch.Tensor([0.0])).cuda()
        recon_loss = gaussian_likelihood(x_hat, log_scale, imgs)
        # kl
        kl = kl_divergence(z, mu, std)
        # elbo
        elbo = (kl - recon_loss)
        elbo = elbo.mean()
        elbo.backward()
        optimizer_encoder.step()
        optimizer_decoder.step()



    if epoch % 20 == 0:
          output2png(x_hat)
          print(elbo)
```
